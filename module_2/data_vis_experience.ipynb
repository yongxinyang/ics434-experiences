```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('default')

fig = plt.figure(figsize=(12,6), facecolor='white')

ax = fig.add_subplot()
ax.set_title("Distribution of Students in the Program", fontdict={'color': 'green'})
ax.set_xlabel("Disciplines")
ax.set_ylabel("Counts")

np.random.seed(42)
# CS”, “Engineering”, “Math” and “Physic
categories = ["CS", "Engineering", "Math", "Physic"]
x = np.random.choice(categories, size=1000)
# print (x)
count_CS = np.count_nonzero(x == 'CS')
count_Eng = np.count_nonzero(x == 'Engineering')
count_Math = np.count_nonzero(x == 'Math')
count_Physic = np.count_nonzero(x == 'Physic')

students_affiliations = (count_Physic, count_CS, count_Math, count_Eng )

# print(students_affiliations)
disciplines = ["Physic","CS", "Math", "Engineering", ]
ax = plt.bar(categories, students_affiliations, color='#FF5733', alpha=0.55)
```
